```python
import pandas as pd
```

```python
df = pd.read_csv('../files/survey_results_public.csv', index_col="Respondent")
```

```python
pd.set_option("display.max_rows", 100)
df["ConvertedComp"]
```

```python
filter_high_salary = df["ConvertedComp"] > 100_000
df.loc[filter_high_salary]
```

```python
df.loc[filter_high_salary].sort_values(by="ConvertedComp", ascending=False).head(10)
```

```python
df.loc[filter_high_salary, ["ConvertedComp", "Age", "Country"]].sort_values(
    by=["ConvertedComp", "Age"], ascending=True, na_position="last"
).head(100)
```

```python
filter_hungary = df["Country"] == "Hungary"
# df.columns
df.loc[
    filter_hungary, ["Age", "DevType", "LanguageWorkedWith", "ConvertedComp"]
].sort_values(by="ConvertedComp", ascending=False).head(100)
```

```python
filter_python = df["LanguageWorkedWith"].str.contains("Python", na=False)
df.loc[filter_python, "LanguageWorkedWith"].sort_values(ascending=False).head(100).str.split(";").str[1].value_counts()
```
